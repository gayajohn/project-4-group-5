```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import make_classification
import pandas as pd
```

```python
data = pd.read_csv('../../Data_df/data_DP3.csv')
data.head(2)
```

```python
data_encoded = pd.get_dummies(data, columns=['industry'])
data_encoded = data_encoded.dropna()
data_encoded
```

```python
y = data_encoded["label"]
target_names = ["increase", "decrease"]
X = data_encoded.drop('label', axis=1)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Initialize the K-Nearest Neighbors classifier with a specified number of neighbors (k)
knn_classifier = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model
knn_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = knn_classifier.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Display classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
