# Data Exploration 
Data exploration is a crucial step in the data analysis process that involves investigating and understanding the characteristics of the dataset.

**Understanding the Data Structure:**
- What are the dimensions of the dataset (rows and columns)?
- What types of variables are present (categorical, numerical, etc.)?

**Identifying Patterns and Trends:**
- Are there any noticeable trends or patterns in the data?
- Are there recurring themes or variations over time?

**Checking Data Quality:**
- Are there missing values in critical fields?
- Are there outliers that may skew the analysis?

**Feature Relationships:**
- How do different features correlate with each other?
- Are there notable dependencies or interactions?

**Supporting Decision-Making:**
- What insights can be gained from the data to inform decisions?
- Are there specific areas that require further investigation?

Descriptive Statistics:

What are the mean, median, and standard deviation of key variables?
What is the distribution of numerical features?

How prevalent are missing values, and in which fields?

````python
# import dependances
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Import seaborn for color palette
import warnings


# Ignore all warnings
warnings.filterwarnings("ignore")
````

````python
# read in data
file_path = 'DB csv files-JB/data_JB11.csv' # REPLACE WITH FINAL DATABASE

df = pd.read_csv(file_path)

df.head()
````

````python
# display the dimensions
print("The stock dataset has", df.shape[0], "records, each with", df.shape[1],
    "attributes")
````

````python
# display the structure
df.info()
````

````python
#Checking for null values
df.isnull().sum()
````

##### The dataset consists of 23 columns, with all columns, except for "industry," containing numerical data. The "label" column serves as a binary classifier with values 0 or 1. Importantly, there are no missing values in the dataset, indicating that all data points across the columns are complete.

````python
# delete rows with empty values
df = df.dropna()
````

````python
# check if there are any null values
df.info()
````

## Evaluate Label Categories
We are dealing with a binary label, where a value of 1 indicates that the stock price increased or remained the same the following day, and a value of 0 signifies that the price decreased on the subsequent day.

````python
df.label.value_counts()
````

##### While the difference between the two classes (class 0 and class 1) is not extreme, the dataset can be considered slightly imbalanced

````python
# plot label distribution
sns.set(style="white")

# Create a bar chart using seaborn with 'rocket' color palette
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='label', data=df, palette='rocket')
ax.set_title("Count of Label")
ax.set_xlabel("Label")
ax.set_ylabel("Count")

# Show the plot
plt.show()
````

## Descriptive Statistics
Now we will use statistical tools to improve the understanding from the dataset, especially the numerical features. We have seen there are 25 numerical columns in the dataset, all of each are financial indicators from the companies. The function ```describe()``` returns information about the distribution of the data such as ```quantiles```, ```min```and ```max```.

````python
# statistical summary
df.describe()
````

````python
# Count the occurrences of each industry
industry_counts = df['industry'].value_counts()

# Sort the counts in descending order
industry_counts = industry_counts.sort_values(ascending=False)

# Define a color palette with as many colors as there are unique industries
colors = sns.color_palette('rocket', len(industry_counts))

# Plot the bar chart with different colors for each bar
plt.figure(figsize=(10, 6))
industry_counts.plot(kind='bar', color=colors)
plt.title('Count by Industry')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
````

## Outliers and Correlations
Identifying outliers and exploring correlations will refine our understanding of the dataset, ensuring data quality and guiding subsequent analysis steps for more accurate and insightful results.

````python
%matplotlib inline

# Assuming 'df' is your DataFrame
# If it's not already loaded, you should load your DataFrame here

# Select columns for the box plots
selected_columns = [col for col in df.columns if col not in ['industry', 'cpi', 'unemployment_rate', 'interest_rate', 'label']]

# Set the style for seaborn
sns.set(style="white")

# Create a box and whisker plot for each feature, separated by label
num_cols = 3
num_rows = -(-len(selected_columns) // num_cols)  # Ceiling division to get the number of rows

plt.figure(figsize=(15, 5 * num_rows))

for i, feature in enumerate(selected_columns, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.boxplot(x='label', y=feature, data=df, palette='rocket')
    plt.title(f'Box and Whisker Plot for {feature}')
    plt.xlabel('Label')
    plt.ylabel(feature)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()

````
